# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Load the dataset

```python
df = pd.read_csv(r'C:\Users\Md Shoaib\c tutorials course\Downloads\AJANTPHARM.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df.head()
```

# Plotting

```python
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Open'], label='Opening Price')
plt.xlabel('Date')
plt.ylabel('Opening Price')
plt.title('Stock Market Direction')
plt.legend()
plt.grid()
plt.show()
```
